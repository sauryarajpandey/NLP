```python
#Text Classification
def gender_features(word):
    return {'last_letter':word[-1]}
```

```python
gender_features('Obama')
```

```python
from nltk.corpus import names
```

```python
names.words('female.txt')
```

```python
print(len(names.words()))
```

```python
labeled_names = ([(name,'male') for name in names.words('male.txt')]+[(name,'female') for name in names.words('female.txt')])
```

```python
print(labeled_names)
```

```python
import random
random.shuffle(labeled_names)
print(labeled_names)

```

```python
featuresets = [(gender_features(n),gender) for (n,gender) in labeled_names]
```

```python
print(featuresets)
```

```python
train_set , test_set = featuresets[:5000], featuresets[2000:]
```

```python
import nltk
classifier = nltk.NaiveBayesClassifier.train(train_set)
```

```python
classifier.classify(gender_features('Saurya raj pandey')
classifier.classify(gender_features('Obama'))
```

```python
print(nltk.classify.accuracy(classifier,test_set)*100)
```

```python
#Vectorization Comparing two texts
from sklearn.feature_extraction.text import CountVectorizer
vect = CountVectorizer(binary=True)
corpus = ["Tesseract is good optical charecter recognition engine","optical charecter recognition is significant"]
```

```python
vect.fit(corpus) #learn unique words. Builds vocab
```

```python
vocab=vect.vocabulary_
```

```python
for key in sorted(vocab.keys()):
    print("{}:{}".format(key,vocab[key]))
```

```python
print(vect.transform(["This is a good optical illusion"]).toarray())
#gives vector: 1 denotes the word present in the sorted corpus
#different corpus gives different vector.
#good corpys is the key
#repeated words does not matter as vector will remain same as it searches unique words
```

```python
#Finding Similarity between documents
from sklearn.metrics.pairwise import cosine_similarity
#far means more angle 
#more similar less angle
#cosine simialrity is for vetorized documnents only
```

```python
similarity = cosine_similarity(vect.transform(["Google Cloud Vision is a charecter recognition engine"]).toarray(),vect.transform(["OCR is an optical charecter recognition engine"]).toarray())
```
